```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
```

```python
bom_df = pd.read_csv('bom.movie_gross.csv.gz')
bom_df
```

```python
bom_df.info
```

```python
bom_df.shape
```

```python
bom_df.columns
```

```python
bom_df = bom_df.dropna(axis=0)
bom_df.info
```

```python
bom_df_without_duplicates = bom_df.drop_duplicates
bom_df
```

```python
income = ['domestic_gross', 'foreign_gross']
bom_df['box office'] = bom_df[income].sum(axis=1)
bom_df
```

top performing films in terms of gross income >$ 100,000,000

```python
bom_df = bom_df[bom_df['domestic_gross'] > 100000000]
bom_df
```

Identify studios used frequently for movie production having domestic and foreign gross of > $100,000,000

```python
bom_df.studio.value_counts()
```

Develop an Histogram to show the freuency of movies used in movie production

```python
Fig, ax = plt.subplots(figsize = (20,8))
ax.hist(bom_df.studio, bins = 80)
ax.set_xlabel('Name of Studio')
ax.set_ylabel('Number of times the studio has been used')
ax.set_title('Studio counts')
```

To know how much the movies made at the box office. The domestic and foreign revenues are added together. The variable box office was formulate

```python
bom_df = bom_df[bom_df['box office'] > 1000000000]
bom_df

```

Plot a bar graph to show the relationship between the movie studion and how much they got at the box office

```python
Fig, ax = plt.subplots(figsize=(15,6))
plt.bar( x= bom_df.studio, height = bom_df.box_office)
plt.xlabel('Name of studios')
plt.ylabel('box office in miliions')
plt.title('Relationship Between studio and total revenue from films')
plt.show()
```
